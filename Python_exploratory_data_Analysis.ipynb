```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
url = '/content/drive/MyDrive/Colab Notebooks/Python_Diwali_Sales_Analysis-main/Diwali Sales Data.csv'
df = pd.read_csv(url,encoding='unicode_escape')
df.shape

```

```python
print(df.head())
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.shape
```

```python
#drop null
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df.info()
```

```python
df['Amount'].dtype
```

```python
#rename colummn
df.rename(columns={'Marital_Status':'Shaadi'})
#rename age
df.rename(columns={'Age':'Umar'})
```

#**Exploratory Data Analysis**

**Gender Count**

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender',data = df)
for bars in ax.containers:
  ax.bar_label(bars)
```

**Grouping by Gender**

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

**Analysis by Age groups**

```python
ax = sns.countplot(data = df,x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

***From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men***

**Checking Amount by Age groups**

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age, hue = 'Age Group')
```

***From above graphs we can see that most of the buyers are of age group between 26-35 yrs female***



**Which state has most orders**

```python
orders_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders', ascending = False)
sns.barplot(y = 'State', x = 'Orders', data = orders_state, hue = 'State')
```

Amountwise States

```python
amt_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(y = 'State', x = 'Amount', data = amt_state, hue = 'State')
```

***From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively***

# **Merital Status**

```python
ax = sns.countplot(data = df,x = 'Marital_Status', hue = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(x= 'Marital_Status', y= 'Amount', data= sales_state, hue= 'Gender')
```

***From above graphs we can see that most of the buyers are married (women) and they have high purchasing power***

```python

ax = sns.countplot(data = df, y = 'Occupation', hue = 'Occupation')
```

```python
occ_sales = df.groupby(['Occupation'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(y = 'Occupation', x = 'Amount', data = occ_sales,hue = 'Occupation')
```

***From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector***

```python
ax = sns.countplot(data = df, y = 'Product_Category',)
```

```python
prod_sales = df.groupby(['Product_Category'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x= 'Amount', y = 'Product_Category', data = prod_sales, hue= 'Product_Category')
```

***From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category***
